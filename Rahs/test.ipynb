```python
import pandas as pd
# excel file
data =  pd.read_excel('/workspaces/AQ-Predators/dataset/Air quality/AAQ Data 2016.xls')
```

```python
data
```

```python
# drop all values till 3rd row and make 4th row as header
data = data.drop(data.index[0:4])
```

```python
# drop unamed0 coloum 
data = data.drop(['Unnamed: 0'], axis=1)
```

```python
# set 1st row as header
data = data.rename(columns=data.iloc[0])
```

```python
# drop 1st row
data = data.drop(data.index[0])
```

```python
# reset index
data = data.reset_index(drop=True)
```

```python
# save as a csv file
data.to_csv('clean AAQ Data 2016.csv', index=False)
```
